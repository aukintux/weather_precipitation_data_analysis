# Data Exploration and Analysis of Monteria Weather Data 

#### Import Modules

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.dates as mdates
%matplotlib inline
```

#### Read CSV into Dataframe

```python
# Variables
years = ["2012", "2013", "2014", "2015"]
df_list = []
# Read in csv files
for year in years:
    temp_df = pd.read_csv("data/weather_mtr_{0}.csv".format(year)) # Read csv file
    temp_df = temp_df[["COT", "Precipitationmm"]] # Filter columns of interest
    temp_df.rename(columns={"COT":"date", "Precipitationmm": "precipitation"}, inplace=True) # Rename column indexes
    temp_df["date"] = pd.to_datetime(temp_df["date"]) # Set proper dtype on date column
    temp_df = temp_df[temp_df["precipitation"] < 200] # Filter out one strangely high precipitation value point
    df_list.append(temp_df) # Append to the dataframes list
# Built dataframe
df = pd.concat(df_list)
df.reset_index(inplace=True)
df.drop("index", axis=1, inplace=True)
df.head()
```

#### Scatter Plot 

```python
# Plot precipitation across all dates available
plt.figure(figsize=(20,5))
plt.scatter(df["date"].tolist(), df["precipitation"].tolist())
plt.xlabel("Date")
plt.ylabel("Precipitation [mm]")
plt.title("Precipitation vs. Date")
plt.show()
```

One can see four (4) main clusters. Each cluster falls in between January - December of each year of data. This mean that there are intervals of time which are located in between the ending and beginning of the years in which rains are scarese.

#### Yearly Scatter Plot

```python
# Construct figure and axis
fig = plt.figure(figsize=(18,10))
axis_list = []
axis_list.append(fig.add_subplot(2,2,1))
axis_list.append(fig.add_subplot(2,2,2))
axis_list.append(fig.add_subplot(2,2,3))
axis_list.append(fig.add_subplot(2,2,4))
# Plot precipitation per year
for index, year_df in enumerate(df_list):
    ax = axis_list[index]
    nonzero_year_df = year_df[year_df["precipitation"] > 0]
    ax.scatter(nonzero_year_df["date"].tolist(), nonzero_year_df["precipitation"].tolist())
    ax.set_title("Year {0}".format(years[index]))
    ax.set_ylim(-10,120)
    ax.set_ylabel("Precipitation [mm]")
    hfmt = mdates.DateFormatter('%b')
    ax.xaxis.set_major_formatter(hfmt)
plt.show()
```

It is hard to identify a pattern to identify rainy months and dry months across all years on first sight. Perhaps a clustering algorithm could do a better job. However one could argue that there is a range of the year, from roughly half of April to November where there is rain. In juxtaposition one also could argue that during a part of December and roughly the first 3.5 months of the year there is almost no rain. Lets explore this assumption further.

#### Number Of Rainy Days Per Year

```python
# Get the number of rainy days per year and the respective percentage
for index, year_df in enumerate(df_list):
    num = sum(year_df["precipitation"] > 0)
    per = num * 100 / year_df.shape[0]
    print("{0} had {1} rainy days which is {2:.2f}% of the year".format(years[index], num, per))
```

One can see that there are rains in about roughly 20% of the year. That mean about 75 rainy days out of the 365 days in a year. 

#### Precipitation Per Week

```python
# Assign week of the year to each row 
for year_df in df_list:
    week_day = 1
    week = 1
    for index, row in year_df.iterrows():
        year_df.loc[index, "week"] = week
        if week_day%7 == 0:
            week_day = 0
            week += 1
        week_day +=1
# Get precipitation per week
weekly_precipitation_list = []
for year_df in df_list:
    weekly_precipitation_list.append(year_df.groupby("week").sum())
# Plot precipitation per week
fig_2 = plt.figure(figsize=(18,10))
axis_list_2 = []
axis_list_2.append(fig_2.add_subplot(2,2,1))
axis_list_2.append(fig_2.add_subplot(2,2,2))
axis_list_2.append(fig_2.add_subplot(2,2,3))
axis_list_2.append(fig_2.add_subplot(2,2,4))
# Plot precipitation per week for each year available
for index, weekly_precipitation in enumerate(weekly_precipitation_list):
    ax = axis_list_2[index]
    ax.bar(weekly_precipitation.index.values.tolist(), weekly_precipitation["precipitation"].tolist())
    ax.set_ylim(0,150)
    ax.set_xlim(1,53)
    ax.set_xlabel("Week of Year")
    ax.set_ylabel("Precipitation [mm]")
    ax.set_title("Year {0}".format(years[index]))
plt.show()
```

The previous plots allow us to see the precipitation for each week of the year, for every year, on a glance it's hard to tell that there are "winter seasons" and "summer seasons". However it appears to be that up to around the 15th week of the year it is indeed very dry. 

#### Rain or No Rain

```python
# Assign week of the year to each row 
for year_df in df_list:
    year_df["rained"] = (year_df["precipitation"] > 0).astype(int)
# Plot rained boolean value
fig_3 = plt.figure(figsize=(18,10))
axis_list_3 = []
axis_list_3.append(fig_3.add_subplot(2,2,1))
axis_list_3.append(fig_3.add_subplot(2,2,2))
axis_list_3.append(fig_3.add_subplot(2,2,3))
axis_list_3.append(fig_3.add_subplot(2,2,4))
# Plot graph showing a line where there was rain
for index, year_df in enumerate(df_list):
    ax = axis_list_3[index]
    ax.bar(year_df.index.values.tolist(), year_df["rained"].tolist())
    ax.set_xlim(1,365)
    ax.set_title("Year {0}".format(years[index]))
    ax.set_ylabel("Rain [0/1]")
    ax.set_xlabel("Day of Year")
# Show plot
plt.show()
```

Further exploration allows us to corroborate that up to around the first 100 days of the year (i.e. the 15th week of the year) the number of rainy days are very scarce. Reason for which is not recommended to plant seeds on this first 100 days. 

#### Duration of Dry Intervals

Check the duration of dry intervals in the year. In this particular case a dry interval is defined as no rain for more than 15 days. This period of time is the number of days that a crop of corn can go without water in this particular region taking into account the soil and luminosity conditions.

```python
# Variable with maximum number of days a crop can go without rain
max_days = 15
# Assign days since the last rain to each row 
for year_df in df_list:
    days_since_last_rain = 0
    for index, row in year_df.iterrows():
        if row["rained"] == 1:
            days_since_last_rain = 0
        else:
            days_since_last_rain += 1
        year_df.loc[index, "days_since_last_rain"] = days_since_last_rain
# Plot days since last rain
fig_3 = plt.figure(figsize=(18,10))
axis_list_3 = []
axis_list_3.append(fig_3.add_subplot(2,2,1))
axis_list_3.append(fig_3.add_subplot(2,2,2))
axis_list_3.append(fig_3.add_subplot(2,2,3))
axis_list_3.append(fig_3.add_subplot(2,2,4))
# Plot each axis 
for index, year_df in enumerate(df_list):
    ax = axis_list_3[index]
    ax.scatter(year_df["date"].tolist(), year_df["days_since_last_rain"].tolist())
    ax.axhline(y=max_days, color='r')
    hfmt = mdates.DateFormatter('%b')
    ax.xaxis.set_major_formatter(hfmt)
    ax.set_ylim(0,150)
    ax.set_ylabel("Number of Days Since Last Rain")
    ax.set_xlabel("Date")
    ax.set_title("Year {0}".format(years[index]))
# Show plot
plt.show()
```

One can see that in particular for a corn crop, which in this particular area accoding to data provided by agronomic engineers can go up to 15 days wihtout water (Specially during the younger stages of the crop). There is relatively good chanches of having a successful crop.
